5. Difícil: Un mapeo interesante a analizar es el mapeo logístico $x_n = r x_{n-1} (1-x_{n-1})$. Al variar el parámetro $r$, los puntos fijos de este mapeo cambian y cambian de estabilidad (de atractivos a repulsivos). Por ejemplo, para $r = 0$ por supuesto el punto $x = 0$ es un punto fijo y es atractivo. Cerca de $x = 0$, el mapeo parece $x_n = r x_{n-1}$ es decir, $r$ define la pendiente de la curva en $x = 0$, así que $x = 0$ es un punto atractivo si $-1<r<1$. Para valores de $r>1$, $x = 0$ sigue siendo un punto fijo, pero deja de ser atractivo. 

    **(i)** Encuentra (analíticamente) qué punto fijo pasa a ser atractivo (es una cuadrática, así que típicamente hay 2 puntos fijos). 
    
    **(ii)** En algún momento, el mapeo pasa a tener una órbita de periodo 2 que es atractiva. Calcula (numéricamente) el valor mínimmo de $r_2>0$ para el cual esto sucede. Recuerda que al aplicar 2 veces el mapeo (para buscar el periodo 2), tendrás una cuártica que tienes que igualar a la identidad, eso te dará 4 soluciones. Tienes que buscar para qué valor de $r$ la pendiente en las dos soluciones que son diferentes de los puntos fijos, tienen valores entre $-1$ y $1$.  
    
    **(iii)** Al seguir incrementando $r$, el mapeo pasa a tener órbitas de cualquier periodo deseado que son atractivas. Haz un código que encuentre el valor mínimo de $r_k$ donde se tiene una órbita atractiva de periodo $k$ cualquiera (es decir, que aplicar k veces el mapeo sobre cualquiera de esos puntos de el mismo valor, pero aplicar $k-1$ veces da un valor diferente. 
    
    **(iv)** Mide la distancia entre los $r$'s donde el perido es $2$ y donde el periodo es $4$. Luego entre $4$ y $8$ y en general mide la distancia entre $r_{2^n}$ para un periodo  $2^n$ y $r_{2^{n+1}}$ para un periodo $2^{n+1}$. 
    
    Nota: El límite $\delta = lim_{n\rightarrow \infty} \frac{r_{2^{n+1}}-r_{2^n}}{r_{2^n}-r_{2^{n-1}}}$ se conoce como la constante de Feigenbaum es una constante universal, es decir, todos los mapeos 1D (y complejos) que tiene bifurcaciones, cumplen con esta constante.
